```python
import tensorflow as tf
import numpy as np
from PIL import Image

# Load the pre-trained model
model1 = tf.keras.models.load_model('gender_model.keras')
```

```python
# Function to preprocess the image
def preprocess_image(image_path, target_size=(56, 56)):
    image = Image.open(image_path)
    image = image.resize(target_size)
    image = np.array(image)
    image = image / 255.0
    image = np.expand_dims(image, axis=0)
    return image
```

```python
# Path to the image you want to predict
image_path = "C:/Users/user/OneDrive/Documents/Age-Gender Prediction/sample images/100_1_2_20170112213615815.jpg.chip.jpg"

# Preprocess the image
preprocessed_image = preprocess_image(image_path)
# Make predictions
prediction = model1.predict(preprocessed_image)
if prediction > 0.5:
    prediction_class=1
else:
    prediction_class=0

# Interpret the results
class_labels = ['male', 'female']  # class 0 is male and class 1 is female

print(f'Predicted class: {class_labels[prediction_class]}')
```
